```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

#Generate a 2D dataset with non-convex clusters


```python
import matplotlib.pyplot as plt
X, y = make_moons(n_samples=300, noise=0.05, random_state=42)

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Visualize
plt.scatter(X_scaled[:, 0], X_scaled[:, 1])
plt.title("Dataset for DBSCAN")
plt.show()
```

#Applying DBSCAN

```python
dbscan = DBSCAN(eps=0.3, min_samples=5)
labels = dbscan.fit_predict(X_scaled)

# Plot clusters
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, cmap='plasma')
plt.title("DBSCAN Clustering")
plt.show()
```

```python
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Extract the feature data
X_iris = iris.data
y_iris = iris.target
```

Standardize the features of the Iris dataset using `StandardScaler` to ensure all features contribute equally to the distance calculation in DBSCAN.


```python
# Instantiate StandardScaler
scaler_iris = StandardScaler()

# Fit and transform the Iris features
X_iris_scaled = scaler_iris.fit_transform(X_iris)
```

```python
# Apply DBSCAN
dbscan_iris = DBSCAN(eps=0.3, min_samples=5)
labels_iris = dbscan_iris.fit_predict(X_iris_scaled)

n_clusters_ = len(set(labels_iris)) - (1 if -1 in labels_iris else 0)
n_noise_ = np.sum(labels_iris == -1)


# Plot clusters
plt.figure(figsize=(8, 6))
plt.scatter(X_iris_scaled[:, 0], X_iris_scaled[:, 1], c=labels_iris, cmap='plasma')
plt.title('DBSCAN Clustering on Iris Dataset (eps=0.3, min_samples=5)')
plt.show()
```
